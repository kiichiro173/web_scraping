## 読売新聞オンラインからHTMLデータの取得

```
url = "https://www.yomiuri.co.jp/"
res = requests.get(url)
soup = BeautifulSoup(res.text,"html.parser")
soup
```

body > div.uni-home-2020.home-2020-wrap > div.home-2020-primebox > div.home-2020-primebox-article > div.home-2020-prime-maincontent > div.home-2020-headline.home-2020-prime-headline > ul.active > li:nth-child(2) > h3 > a

```
elems = soup.select("ul")
elems
```

```
soup.find_all("h3")
```

```
elems_news = soup.find_all("h3")
```

```
for elem in elems_news[4:]:
    print(elem.a.string)
    print(elem.a["href"] , '\n')
```

# yahooから主要トピックをスクレイピング

```
url = "https://www.yahoo.co.jp/"
res = requests.get(url)
soup = BeautifulSoup(res.text , "html.parser")
soup
```

```
elems = soup.find_all("a")
elems
```

# 正規表現を使用して主要なトピックを取得している。

```
import re 
elems = soup.find_all(href=re.compile("news.yahoo.co.jp/pickup"))
elems
```

```
#主要トピックを取得
for elem in elems:
    print(elem.span.string)
    print(elem["href"] , "\n")
```
