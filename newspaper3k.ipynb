### ブルームバーグから記事を取得していく

```
url = "https://www.bloomberg.co.jp/"
```

```
#buildを使用することによってサイトのトップページから複数の記事を取得することが出来る。
website = newspaper.build(url)
```

```
#複数の記事のリストを返す。
website.articles
```

```
i = 0
for article in website.articles:
    #まずは分析を行う準備
    article.download()
    article.parse()
    article.nlp()#自然言語処理

    print(article.title)
    print(article.url)
    print(article.summary , end='\n\n')

    if i > 9:
        break
    i += 1
```
