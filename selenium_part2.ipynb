```python
from bs4 import BeautifulSoup
#投稿件数を取得する。
try:
    #textに含まれている文字を使用する場合はtext()と入力をする。
    post_count = driver.find_element_by_xpath("//span[text() = '投稿']").text
    #件数のみを取得するためreplaceを使用する
    post_count = post_count.replace('投稿' , '').replace('件' , '')
    print("投稿件数;" + post_count)
    post_count = int(post_count)
    #インスタグラムでは12枚の画像が表示されたら次の画像を読み込む。
    #従って投稿件数が12を超えたらスクロール回数を計算する。
    if post_count > 12:
        #スクロール回数を計算する。
        scroll_count = int(post_count/12) + 1
        
        #スクロールの回数だけ繰り返し処理を行っていく。
        try:
            all_images = []#この中にはimageが重複している。
            #for文の中で画像を取得するようにしていく。
            #seleniumでソースコードを取得するにはdriver.page_sourceで行っていく。
            #画面をスクロールする前にHTMLコードを取得してそこからIMGタグのSRCを取得する。
            #画像を取得する際はスクロールしながら出ないと取得することが出来ない、
            #スクロールした後にスクレイピングを行ったとしても一部取得出来ないといったことがあるのでスクロールしながら出ないといけない。
            for i in range(scroll_count):
                #画面のスクロールを行っている。
                soup = BeautifulSoup(driver.page_source , "html.parser")#ここでBeautifulSoupに読み込んでいる。
                #imgタグのついているものを取得している。
                for image in soup.find_all('img'):
                    all_images.append(image)
                driver.execute_script('window.scrollTo(0,document.body.scrollHeight);')
                #スクロールが5回行われたら終わり
                if i > 5:
                    break
            
            #リストの中の重複した分を取り除いていく。
            all_images = list(dict.fromkeys(all_images))
            
            
            for index , image in enumerate(all_images):
                print("画像番号:" + str(index))
                print("image['src']:" + image['src'] , end = '\n\n')
                
        except Exception:
            print("画面スクロール中にエラーが発生しました。")
            error_flg = True

        

except Exception:
    print("投稿数が取得出来ませんでした")
    error_flg = True
        
```

```python
import requests
import os
import re
import shutil

#画像を保存するパスを指定する。
path = r"C:\instagram"

if error_flg is False:
    try:
        
        for index , image in enumerate(all_images):
            filename = 'image_' + str(index) + '.jpg'
            #複数のパスの要素を結合する→os.path.join(ファイルへのパス,ファイル名)
            image_path = os.path.join(path , filename)
            image_link = image["src"]#画像のデータが入っている。

            #正規表現を使用してこのパターンがマッチしているかどうかを確認します。
            url_ptn = re.compile(r"'http|https://'")
            res = url_ptn.match(image_link)

            #画像をダウンロードする。
            if res:
                response = requests.get(image_link , stream=True)#image_linkに画像のリンクが入っている。
                with open(image_path , 'wb') as file:
                    #ファイル形式のオブジェクトをファイルへコピーする
                    shutil.copyfileobj(response.raw , file)
    except Exception as e:
        print(e)
        print(str(index) + "番目の画像ダウンロード・保存時にエラーが発生しました。")
        print("画像へのリンク:" + image_link)
```
