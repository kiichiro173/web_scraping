```
#Beautiful Soupインポート
from bs4 import BeautifulSoup
import requests
```

```
html_doc = """
<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>
<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>
<p class="story">...</p>
"""

# BeautifulSoupの初期化

soup = BeautifulSoup(html_doc, 'html.parser') # BeautifulSoupの初期化
```

```
soup
```

```
#aタグを一行だけ返す
soup.a
```

```
#selectを行うと()の中にあるタグをすべて取得することが出来る。
soup.select("a")
```

```
#findはタグを一つ見つけてくる。
soup.find("body")
```

```
#htmlをインデントすることが出来る
print(soup.prettify())
```

```
#タイトルを取得する。(タグつき)
soup.find("title")
```

```
#タイトルを取得する。
soup.find("title").string
```

```
#find_allを使用するとリスト形式で指定したタグを取得することが出来る。
soup.find_all("a")
```

```
#aタグにあるUrlと名前を取得する。
url_list = []
string_list = []
for tag in soup.select('a'):
    if tag != '\n':
        print(tag["href"])
        print(tag.string)
        url_list.append(tag["href"])
        string_list.append(tag.string)
```

```
#getメゾットを使用する事でtagの中にあるものを取得することも可能
tags = soup.find_all('a')
for tag in tags:
    print(tag.get("href"))
```

```
#urlからHTMLを取得するためにrequestsをimport
import requests
url ="https://review-of-my-life.blogspot.com/"
res = requests.get(url)
print(res.text)
```

```
soup = BeautifulSoup(res.text , "html.parser")
print(soup.prettify())
```

```
#aタグの文字だけを取得する。
a_tags = soup.find_all('a')
for tag in a_tags:
    print(tag.string)
```

```
#リンクを取得
for tag in a_tags:
    print(tag.get("href"))
```

# 特定のクラスを取得したいとき
find_get()メゾットを使用する<br>
具体的なこととしては<br>
soup.find_all("h3" , {"class":"post-title"})<br>
これはh3タグのclassがpost-titleのものを取得という意味


```
h3_tags = soup.find_all("h3" , {"class":"post-title entry-title"})
h3_tags
```

```
#タイトルとURLを取得
for h3_tag in h3_tags:
    print(h3_tag.a.string)
    print(h3_tag.a.get("href"))
```

```
import pandas as pd

columns = ["name" , "url"]
df = pd.DataFrame(columns=columns)
df
```

```
#タイトルとURLを取得しデータフレームに追加
for h3_tag in h3_tags:
    se = pd.Series([h3_tag.a.string,h3_tag.a.get("href")] , columns )
    df = df.append(se , columns)

df
```
