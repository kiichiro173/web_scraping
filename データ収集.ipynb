```
from bs4 import BeautifulSoup
import requests
'''
requestsで情報を持ってくる。
beautifulsoupでパースをする。
パースとは解析って意味
'''
```

```
url = 'https://scraping-for-beginner.herokuapp.com/udemy'
```

```

r = requests.get(url)
```

```
r.text
```

```
#構造の解析を行っている。
soup = BeautifulSoup(r.text , 'html.parser')
```

```
soup
```

```
#.(ドット)を忘れずに！
#リストで返ってくる
soup.select('.card-title')
```

```
soup.select('.card-title')[0].string
```

```
name = soup.select(".subscribers")[0].string

print(name.split('：')[1])
```

```
'''
マッチする部分があるとre.search()はマッチオブジェクトを返す。マッチオブジェクトのgroup()メソッドでマッチした部分を文字列として取得できる。
'''
import re
s = soup.select(".subscribers")[0].string
m = re.search(r'[0-9]+', s)

print(m.group())

```
