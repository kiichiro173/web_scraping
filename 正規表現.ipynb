```
import re
url = "http://www.amazon.co.jp/dp/B07T9TCPZX"
```

```
pattern = "https?://[^/]+/"
```

```
#machは先頭からパターンがマッチしているのかを表示する。
m = re.match(pattern , url)
#マッチした文字を取得する。
m.group()
```

```
#マッチした開始位置と終了位置を表示している。
m.span()
```

```
#re.match→先頭の文字列からパターンに一致するものを検索
m = re.match("x.z" , "xyz")
m
```

```
m = re.match("x.z" , "xxyz")
print(m)
```

```
#searchは先頭に限らずパターンに一致するものを検索
#ただし複数一致したとしても一つ目だけを返す。

m = re.search("x.z" , "xxyz")
#さっきはNoneであったのに対し、今回はしっかりと出来ていることが分かる。
print(m)
```

```
#パターンに一致するものは複数あるが、一個しか取得されない→複数取得したい場合はfindallを使用する。
m = re.search("x.z" , "xxyz　xxyz")
print(m)
```

```
#findallを使用することによってパターンに一致したものすべてを取得している。
m = re.findall("x.z" , "xxyz　xxyz")
print(m)
```

```
#finditerを使用することによって、パターンに一致するものをすべてmatchオブジェクトで取得している。
#戻り値はイテレータを返している。
#for分を回して取得する必要がある。
re.finditer("x.z" , "xxyz　xxyz")
```

```
#こんな感じでfor文を回すことによってすべてmatchオブジェクトで取得することが出来る。
for s in re.finditer("x.z" , "xxyz　xxyz"):
    print(s)
```

```
#文字全体が一致しているかどうかを確認する。
m = re.fullmatch("x.z" , "xxyzxxyz")
print(m)
```

```
#すべて一致するやつのみ返り値が出てくる。
m = re.fullmatch("x.*z" , "xxyzxxyz")
print(m)
```

```
#subはパターンに一致した文字列を別の文字列に置き換える。
#マッチしなければそのまま文字列が返ってくる。
#re.sub(パターン,置換する文字列,検索する文字列)
m = re.sub("x.z" , "abc" , "vwxyz")
print(m)
```

# patternを使用した正規表現

```
pattern = re.compile("xy+")
m = pattern.match("xyyyyyyy")
print(m)
```
