```
import pandas as pd
from bs4 import BeautifulSoup 
import requests
```

# 研音アーティスト
名前・プロヒールURL・画像URL取得

```
url = "https://www.ken-on.co.jp/artists/"
res = requests.get(url)
soup = BeautifulSoup(res.content , "html.parser")
soup
```

```
artist_lists = soup.find_all("ul" , {"class":"artist__list"})
artist_lists[4]
```

```
print(artist_list[4].a.get("href"))#アーティスト個別ページ
print(artist_list[4].img.get("src"))#写真
artist_list[4].p.string#アーティスト名

```

```
#アーティストの名前、イメージ画像、パーソナル紹介サイトの取得
artist_name = []
artist_img = []
artist_page = []
artist_feature = []
for artist_list in artist_lists:
    for person in artist_list.find_all('li'):
        artist_name.append(person.p.string)
        artist_img.append(person.img.get("src"))
        artist_page.append(person.a.get("href"))

        


```

```
#すべて取得しているのかを確認する。
print(len(artist_name))
print(len(artist_img))
print(len(artist_page))
```

```
#データフレームにする。
df = pd.DataFrame()
df["artists"] = artist_name
df["images"] = artist_img
df["person_url"] = artist_page
df.head()
```
