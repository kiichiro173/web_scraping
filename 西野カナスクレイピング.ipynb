```
import pandas as pd
from bs4 import BeautifulSoup 
import requests
```

```
url = "https://www.uta-net.com/artist/7429/"#西野カナのURL
base_url = 'https://www.uta-net.com'
res = requests.get(url)
soup = BeautifulSoup(res.content , "html.parser")
```

```
#歌のタイトルと歌詞のURLが入っているタブを取得 
song_list = soup.find_all("td" , {"class":"side td1"})
```

```
song_list[0]
```

```
song_list[0].a["href"]
```

```
song_list[0].a.string
```

```
song_title_list = []
song_url_list = []

for song in song_list:
    song_title_list.append(song.a.string)
    song_url = base_url + song.a["href"]
    song_url_list.append(song_url)
```

```
kana_df = pd.DataFrame({"Title":song_title_list , "URL":song_url_list})
kana_df
```

```
res = requests.get(kana_df["URL"][0])
soup2 = BeautifulSoup(res.content , "html.parser")
```

```
soup2.find_all("div" ,{"itemprop":"text"})[0].text
```

```
escape = str.maketrans({"\n":'',"\t":'' , "\u3000":' '})
soup2.find_all("div" ,{"itemprop":"text"})[0].text.translate(escape)
```

```
#歌詞を入力していく。
lyric_lists = []
escape = str.maketrans({"\n":'',"\t":'' , "\u3000":' '})

for kana_url in kana_df["URL"]:
    res = requests.get(kana_url)
    soup2 = BeautifulSoup(res.content , "html.parser")
    lyric = soup2.find_all("div" ,{"itemprop":"text"})[0].text.translate(escape)
    lyric_lists.append(lyric)
print(len(lyric_lists))
```

```
kana_df["text"] = lyric_lists
kana_df.head()
```
