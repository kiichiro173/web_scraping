## タイトル・著者・掲載媒体名を持ってくる。

```
main_contents = soup.find_all("div", {"class":"maincontents"})
main_contents
```

```
#タイトルの取得
a = soup.find_all('dt' , {'class':"item_mainTitle item_title"})
count = 0
for i in a:
    escape = str.maketrans({"\n":'',"\t":''})
    count += 1
    print(count)
    if i.text == None:
        print(i.string)
    else:
        print(i.text.translate(escape))
    
```

```
#著作者の取得
a = soup.find_all('p' , {'class':"item_subData item_authordata"})
count = 0
for i in a:
    escape = str.maketrans({"\n":'',"\t":''})#ここで空白文字の削除を行っている。
    count += 1
    print(count)
    print(i.text.translate(escape))
    
   
```

```
#掲載媒体の取得
a = soup.find_all('p' , {'class':"item_extraData item_journaldata"})
count = 0
for i in a:
    escape = str.maketrans({"\n":'',"\t":''})#ここで空白文字の削除を行っている。
    count += 1
    print(count)
    print(i.text.translate(escape))
    
```

```
#著者・タイトル・掲載媒体を一気に取得
title_list = []
author_list = []
media_list = []

#ここで空白文字の削除を行っている。
escape = str.maketrans({"\n":'',"\t":''})
count = 0
for paper in soup.find_all("dl", {"class":"paper_class"}):#1つの論文ごとにループを回す。
    count += 1
    #タイトルの取得
    title_list.append(paper.a.text.translate(escape))
    #著者の取得
    author_list.append(paper.find('p' , {'class':"item_subData item_authordata"}).text.translate(escape))
    #掲載媒体の取得
    media_list.append(paper.find('p' , {'class':"item_extraData item_journaldata"}).text.translate(escape))
 
print(len(title_list))
print(len(author_list))
print(len(media_list))
```

```
pd.DataFrame({"Title":title_list , "Author":author_list , "Media":media_list})
```

```
def inami(soups):
    #著者・タイトル・掲載媒体を一気に取得
    title_list = []
    author_list = []
    media_list = []

    #ここで空白文字の削除を行うための処理
    escape = str.maketrans({"\n":'',"\t":''})
    
    for paper in soups.find_all("dl", {"class":"paper_class"}):#1つの論文ごとにループを回す。
        
        #タイトルの取得
        title_list.append(paper.a.text.translate(escape))
        #著者の取得
        author_list.append(paper.find('p' , {'class':"item_subData item_authordata"}).text.translate(escape))
        #掲載媒体の取得
        media_list.append(paper.find('p' , {'class':"item_extraData item_journaldata"}).text.translate(escape))
    
    return title_list ,  author_list,media_list
```
